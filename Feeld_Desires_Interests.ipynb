### splitting dataframes into different demographics

```python

straight_woman_desires_df = desires_df[(desires_df['gender']=='woman') & (desires_df['orientation']=='straight')]
straight_woman_interests_df = interests_df[(df['gender']=='woman') & (df['orientation']=='straight')]

bisexual_woman_desires_df = desires_df[(df['gender']=='woman') & (df['orientation']=='bisexual')]
bisexual_woman_interests_df = interests_df[(df['gender']=='woman') & (df['orientation']=='bisexual')]

queer_woman_desires_df = desires_df[(df['gender']=='woman') & (df['orientation']=='queer')]
queer_woman_interests_df = interests_df[(df['gender']=='woman') & (df['orientation']=='queer')]


gay_man_desires_df = desires_df[(df['gender']=='man') & (df['orientation']=='gay')]
gay_man_interests_df = interests_df[(df['gender']=='man') & (df['orientation']=='gay')]

bisexual_man_desires_df = desires_df[(df['gender']=='man') & (df['orientation']=='bisexual')]
bisexual_man_interests_df = interests_df[(df['gender']=='man') & (df['orientation']=='bisexual')]

queer_man_desires_df = desires_df[(df['gender']=='man') & (df['orientation']=='queer')]
queer_man_interests_df = interests_df[(df['gender']=='man') & (df['orientation']=='queer')]
```

### creating function to calculate top 5 desires and interests

```python
def top(df):
    top_dict = {}
    counts = df.sum(numeric_only=True)
    names = counts.index
    totals = []
    for name,count in zip(names,counts):
        if 'desire:' in name:
            top_dict[name[8:]] = round(count/len(df),2)
        elif 'interest:' in name:
            top_dict[name[9:]] = round(count/len(df),2)
        totals.append(count)
    d_or_i = top_dict.keys()
    percents = top_dict.values()
    if 'desire:' in names[0]:
        results = pd.DataFrame({'desire':d_or_i,'percentage':percents, 'counts':totals})
    elif 'interest' in names[0]:
        results = pd.DataFrame({'interest':d_or_i,'percentage':percents, 'counts':totals})
    results = results.sort_values(by='percentage',ascending=False)
    return results[:5]
                        
```

### applying function to different demographics

```python
top_all_desires_df = top(desires_df)
top_all_desires_df
```

```python
top_all_interests_df = top(interests_df)
top_all_interests_df
```

```python

top_straight_woman_desires_df = top(straight_woman_desires_df)
top_straight_woman_interests_df = top(straight_woman_interests_df)

top_bisexual_woman_desires_df = top(bisexual_woman_desires_df)
top_bisexual_woman_interests_df = top(bisexual_woman_interests_df)

top_queer_woman_desires_df = top(queer_woman_desires_df)
top_queer_woman_interests_df = top(queer_woman_interests_df)



top_gay_man_desires_df = top(gay_man_desires_df)
top_gay_man_interests_df = top(gay_man_interests_df)

top_bisexual_man_desires_df = top(bisexual_man_desires_df)
top_bisexual_man_interests_df = top(bisexual_man_interests_df)

top_queer_man_desires_df = top(queer_man_desires_df)
top_queer_man_interests_df = top(queer_man_interests_df)
```
